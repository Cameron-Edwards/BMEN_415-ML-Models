Classification data(Breast-Cancer dataset)

```python
import os
import numpy as np 
import pandas as pd 
import seaborn as sns 
from sklearn import datasets
from sklearn import tree
from sklearn import svm
from sklearn import metrics
import matplotlib.pyplot as plt 
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
breast_cancer_data = pd.read_csv("breast-cancer.csv")
breast_cancer_data
```

```python
breast_cancer_data.drop('id', axis=1, inplace= True)
breast_cancer_data.diagnosis = [1 if each == 'M' else 0 for each in breast_cancer_data.diagnosis]
breast_cancer_data
```

```python
breast_cancer_data['diagnosis'].unique()
```

```python
breast_cancer_data.describe()
```

```python
features = ['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean',
        'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']
```

```python
Y = breast_cancer_data.diagnosis
X = breast_cancer_data.drop('diagnosis', axis=1)

X.shape, Y.shape
print(type(X))

X_numpy=X.to_numpy()
Y_numpy=Y.to_numpy()

print(type(X_numpy))
print(X_numpy.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_numpy, Y_numpy, test_size=0.2)
```

Model 1: Suppor Vector Machines (SVM)

```python
model_1 = svm.SVC(kernel="linear")         
model_1.fit(X_train, Y_train)               
y_pred_test = model_1.predict(X_test)
y_pred_train = model_1.predict(X_train)              
print('Training accuracy :', model_1.score(X_train,Y_train))
print('Testing accuracy :', model_1.score(X_test, Y_test))
print('Model accuracy: ', accuracy_score(Y_test, y_pred_test))
print("precision:", metrics.precision_score(Y_test, y_pred_test))
print(classification_report(Y_test, y_pred_test, target_names= ['Malignant', 'Benign']))
```

```python
confusion_matrix(Y_test,y_pred_test, labels=[1, 0])
```

```python
from sklearn.metrics import plot_confusion_matrix

confusion_matrix1 = confusion_matrix( Y_test,y_pred_test)

cm1 = sns.heatmap(confusion_matrix1, square=True, annot=True, fmt='d', cbar=False)
cm1.set_xlabel('\nPredicted Values')
cm1.set_ylabel('Actual Values')
cm1.xaxis.set_ticklabels(['Benign', 'Malignant'])
cm1.yaxis.set_ticklabels(['Benign', 'Malignant'])
plt.show()
```

Model 2: Logistic Regression Model

```python
model_2 = LogisticRegression(max_iter=20000)
model_2.fit(X_train,Y_train)
Prediction_2 = model_2.predict(X_test)

print('Training accuracy :', model_2.score(X_train,Y_train))
print('Testing accuracy :', model_2.score(X_test, Y_test))
print('Model accuracy: ', accuracy_score(Y_test, Prediction_2))
print("precision:", metrics.precision_score(Y_test, Prediction_2))
print(classification_report(Y_test, Prediction_2, target_names= ['Malignant', 'Benign']))
```

```python
confusion_matrix(Y_test,Prediction_2, labels=[1, 0])
```

```python
confusion_matrix_2 = confusion_matrix(Y_test, Prediction_2)
sns.heatmap(confusion_matrix_2, square=True, annot=True, fmt='d', cbar=False)
plt.show()
```

Model 3: Basic CART Classification Model

```python
model_3 = DecisionTreeClassifier(ccp_alpha=0.005)
model_3 = model_3.fit(X_train, Y_train)
Prediction_3 = model_3.predict(X_test)
```

```python

print('Training accuracy :', model_3.score(X_train,Y_train))
print('Testing accuracy :', model_3.score(X_test, Y_test))

print('Model accuracy: ', accuracy_score(Y_test, Prediction_3))
print("precision:", metrics.precision_score(Y_test, Prediction_3))
print(classification_report(Y_test, Prediction_3, target_names= ['Malignant', 'Benign']))
```

```python
confusion_matrix(Y_test,Prediction_3, labels=[1, 0])
```

```python
confusion_matrix_3 = confusion_matrix(Y_test, Prediction_3)
sns.heatmap(confusion_matrix_3, square=True, annot=True, fmt='d', cbar=False)
plt.show()
```

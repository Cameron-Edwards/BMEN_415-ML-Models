```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt  
import seaborn as sns 
import os
from sklearn import tree
from sklearn import datasets
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_absolute_error

```

```python
data = pd.read_csv("C:/Users/youss/Desktop/Bmen 415 project/Volumetric_features.csv")
data
```

```python
data.info()
data.shape
```

```python
data.drop('S.No', axis=1, inplace=True)
data.drop('dataset', axis = 1, inplace=True)
```

```python
data.describe()
```

```python
data.corr()
```

```python
Y = data.Age
X = data.drop('Age', axis=1)
X.shape, Y.shape
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size= 0.2)

x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.datasets import make_regression
modelADA = AdaBoostRegressor(random_state=20, n_estimators=50)
modelADA.fit(x_train, y_train)
predictions_ADA = modelADA.predict(x_test)
print(predictions_ADA)
```

```python
from sklearn.metrics import r2_score  # For find accuracy with R2 Score
from sklearn.metrics import mean_squared_error  # For MSE
from math import sqrt  # For squareroot operation

MSE_ADA = mean_squared_error(y_test, predictions_ADA)
print("MSE for Testing Data: ", MSE_ADA)

RMSE_ADA= sqrt(MSE_ADA)
print("RMSE for Testing Data: ", RMSE_ADA)

print("Mean Absolute error for Testing Data", mean_absolute_error(y_test, predictions_ADA))
print('Coefficient of determination (R^2): %.2f', r2_score(y_test, predictions_ADA))
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(y_test, predictions_ADA, color='orange')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', color='green', lw = 3)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual Vs Predicted value')
```

Model 2: Decision Tree Regression

```python
modelDT = DecisionTreeRegressor(random_state=20)
modelDT.fit(x_train, y_train)
predictions = modelDT.predict(x_test)
print(predictions)
```

```python
from sklearn.metrics import r2_score  # For find accuracy with R2 Score
from sklearn.metrics import mean_squared_error  # For MSE
from math import sqrt  # For squareroot operation

MSE_DT = mean_squared_error(y_test, predictions)
print("MSE for Testing Data: ", MSE_DT)

RMSE_DT= sqrt(MSE_DT)
print("RMSE for Testing Data: ", RMSE_DT)

print("Mean Absolute error for Testing Data", predictions)
print('Coefficient of determination (R^2): %.2f', r2_score(y_test, predictions))
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(y_test, predictions, color='grey')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', color='blue', lw = 4)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual Vs Predicted value')

```

Model 3: Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

model_LinearReg = LinearRegression(fit_intercept=True)
model_LinearReg.fit(x_train, y_train)
y_pred_train_LinearReg = model_LinearReg.predict(x_train)
y_pred_test_LinearReg = model_LinearReg.predict(x_test)
```

```python
from sklearn.model_selection import cross_val_predict  # For K-Fold Cross Validation
from sklearn.metrics import r2_score  # For find accuracy with R2 Score
from sklearn.metrics import mean_squared_error  # For MSE
from math import sqrt  # For squareroot operation

accuracy_train_LinearReg = model_LinearReg.score(x_train, y_train)
print("Training Accuracy: ", accuracy_train_LinearReg)

accuracy_test_LinearReg = model_LinearReg.score(x_test, y_test)
print("Testing Accuracy: ", accuracy_test_LinearReg)

RMSE_train_LinearReg = sqrt(mean_squared_error(y_train, y_pred_train_LinearReg))
print("RMSE for Training Data: ", RMSE_train_LinearReg)

RMSE_test_LinearReg = sqrt(mean_squared_error(y_test, y_pred_test_LinearReg))
print("RMSE for Testing Data: ", RMSE_test_LinearReg)
```

```python
print('Coefficients:', model_LinearReg.coef_)
print('Intercept:', model_LinearReg.intercept_)
print('Mean Squared error (MSE): %.2f', mean_squared_error(y_test, y_pred_test_LinearReg))
print('Coefficient of determination (R^2): %.2f', r2_score(y_test, y_pred_test_LinearReg))
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(y_train, y_pred_train_LinearReg, color='purple')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'k--', color='black', lw = 4)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual Vs Predicted value')
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(y_test, y_pred_test_LinearReg, alpha=1.0, marker="*")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', color='blue', lw = 4)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual Vs Predicted value')
```

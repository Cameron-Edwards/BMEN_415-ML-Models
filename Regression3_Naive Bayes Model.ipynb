```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt  
import seaborn as sns 
import os
from sklearn import datasets
from sklearn.metrics import confusion_matrix, mean_squared_error, mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_predict  
from math import sqrt  # For squareroot operation
```

```python
data = pd.read_csv("vol_feature.csv")
data
```

```python
data.info()
data.shape
```

```python
data.drop('S.No', axis=1, inplace=True)
data.drop('dataset', axis = 1, inplace=True)
```

```python
data.describe()
```

```python
data.corr()
```

```python
Y = data.Age
X = data.drop('Age', axis=1)
X_numpy=X.to_numpy()
Y_numpy=Y.to_numpy()
X.shape, Y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_numpy,Y_numpy, test_size= 0.2)

x_train.shape, y_train.shape
```

```python
from sklearn.naive_bayes import GaussianNB
model_3 = GaussianNB()
model_3.fit(x_train, y_train)
Prediction_3 = model_3.predict(x_test)
```

```python
print('MSE:', mean_squared_error(y_test, Prediction_3))
print('MAE:', mean_absolute_error(y_test, Prediction_3))

print("RMSE for Training Data: ", sqrt(mean_squared_error(y_train, model_3.predict(x_train))))
print("RMSE for Testing Data: ", sqrt(mean_squared_error(y_test, Prediction_3)))

print('Coefficient of determination (R^2): %.2f'
        % r2_score(y_test, Prediction_3))
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(y_test, Prediction_3, alpha=1.0, marker="*")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', color='red', lw = 3)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual Vs Predicted value')
```
